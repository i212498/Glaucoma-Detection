```python
import cv2

def convert_tiff_to_png(input_file, output_file):
    # Read TIFF image
    tiff_img = cv2.imread(input_file, cv2.IMREAD_UNCHANGED)

    # Check if the TIFF image is loaded successfully
    if tiff_img is None:
        print("Error: Unable to read TIFF image.")
        return
    
    tiff_img[tiff_img > 0] = 255
    
    

    # Write PNG image
    cv2.imwrite(output_file, tiff_img)

    print(f"Conversion completed: {input_file} -> {output_file}")


for i in range(1, 101):
    input_file = ".tif"
    output_file = ".png"
    convert_tiff_to_png(str(i) + input_file, str(i) + output_file)
```
